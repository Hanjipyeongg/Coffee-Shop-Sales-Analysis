```python
import pandas as pd
```

```python
df = pd.read_csv('coffeeshop_dataset.csv')
df.head()

```

```python
#dataset information and statistical summary 
df.info()
df.describe(include='all')
```

```python
#unique value exploration
payment_type = df['cash_type'].unique()
list_menu = df['coffee_name'].unique()
print("list cash:", payment_type)
print("list_name:", list_menu)
```

```python
#check duplicate
total_duplicate = df.duplicated().sum()
print(total_duplicate)
```

```python
#delete column date and check detail information from my dataset
df = df.drop(columns=['date'])
df.info()
```

```python
#change format column datetime & rename column
df['datetime'] = pd.to_datetime(df['datetime'])
df.rename(columns={'datetime': 'date'}, inplace=True)
df.info()
df.head(10)
```

```python
#rename column money to total_amount and change the capitalization format in the dataset
df.rename(columns={'money': 'total_amount'}, inplace=True)
df[['cash_type','card']] =df[['cash_type', 'card']].apply(lambda x: x.str.title())
df.head(10)
```

```python
#checking null value & count total null value
df_null = df[df['card'].isnull()]
print(df.isnull().sum())
```

```python
#change null values from card column to 0 because it's count as a real transaction
df['card'] = df['card'].fillna('Cash')
df.info()
df.head(10)
```

```python
#Export data

df.to_csv('coffeeshop_dataset_clean.csv', index=False)
print("Export data success!")
```
